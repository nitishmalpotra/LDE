### Some questions / queries
- What is the max and min Electric Range in these EVs?
- what is the name of the EVs with max and min Range?
- List all cars with range > 300 KM
- create a column with EV Make - Model - Year
- Sort cars by Model Year and range, return latest cars (top 5)

```python
# What is the max and min Electric Range in these EVs?
print("Max range is " +
      str(df['Electric Range'].max()) +
      " km\nand min range is " +
      str(df['Electric Range'].min()) + " km")
```

```python
# what is the name of the EVs with max Range?
df[df['Electric Range'] == 337]
```

```python
# what is the name of the EVs with min Range?
df[df['Electric Range'] == 20]
```

```python
# List all cars with range > 200 KM
df[df['Electric Range'] > 300]
```

```python
# create a column with EV Make - Model - Year
df["EV Full Name"] = df["Make"] + " - " + df["Model"] + " - " + df["Model Year"].astype(str)
df.head()
```

```python
# Sort cars by Model Year and range, return latest cars (top 6)
df.sort_values(by=['Model Year', 'Electric Range'], ascending = False)[:7]
```
